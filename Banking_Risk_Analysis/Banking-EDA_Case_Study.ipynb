```python
bins = [0, 100000, 300000,float('inf')]
labels = ['low', 'medium', 'high']

# Create income band column
df['income_band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels)

df.head()
```

```python
df['income_band'].value_counts().plot(kind='bar')
```

```python
#examine the distribution of unique categories in categorical columns
categorical_cols=df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","income_band",]].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Loop through each categorical column and plot
for col in categorical_cols:
    plt.figure(figsize=(10, 5))
    ax = sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='viridis')
    
    plt.title(f"Distribution of {col}", fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.xlabel(col)
    plt.ylabel("Count")
    
    # Add count labels on top of bars
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2., height + 3, f'{int(height)}', ha="center")

    plt.tight_layout()
    plt.show()
```

```python
for col in categorical_cols:
    if col != "Amount of Credit Cards":
        sns.countplot(data=df, x=col, hue="Amount of Credit Cards", order=df[col].value_counts().index); plt.title(f"{col} by Credit Cards"); plt.xticks(rotation=45); plt.tight_layout(); plt.show()
```

```python
top_occupations = df["Occupation"].value_counts().nlargest(10).index

plt.figure(figsize=(10, 5))
sns.countplot(data=df[df["Occupation"].isin(top_occupations)],
              x="Occupation",
              order=top_occupations,
              palette="muted")

plt.title("Top 10 Occupations")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Histplot of value counts for different occupations

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

### Numerical Analysis

```python
# Automatically fetch all numerical columns
numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Plot histograms for each numerical column
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

### Heatmaps

```python
# Select only numerical columns
numerical_cols = df.select_dtypes(include=["int64", "float64"])

# Compute correlation matrix
corr_matrix = numerical_cols.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="crest", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.show()
```

##### High deposits correlate strongly with savings and checking accounts
(Bank Deposits ↔ Saving Accounts = 0.75, Checking Accounts = 0.84)	Customers who maintain high balances in their bank deposits also tend to have higher balances in savings and checking accounts.

📌 This indicates a segment of financially responsible users who are good at managing liquid assets.

✅ Actionable: These customers can be targeted for premium savings products, investment services, or priority banking offers.
2. Estimated Income correlates positively with Risk Weighting and Credit Access
(Estimated Income ↔ Risk Weighting = 0.66, ↔ Amount of Credit Cards = 0.37)	Higher income customers are more likely to have multiple credit cards and higher risk exposure. This may be because they are offered more credit or engage in more complex financial behaviors.

📌 While they are valuable, they may also pose higher potential exposure in risk models if balances are not well managed.

✅ Actionable: These individuals should be carefully monitored using credit limit optimization, and risk-adjusted pricing strategies may be applied.
3. Bank Deposits serve as a strong proxy for financial strength
(Bank Deposits has strong correlations with many financial variables: Checking, Savings, Loans, etc.)	The amount held in bank deposits appears to be a central indicator of a customer’s financial capability and engagement.

📌 It could be used as a key feature in a credit scoring or churn prediction model.

✅ Actionable: Use deposit size to segment customers into tiers and offer tier-based loyalty or financial planning tools.
4. Age is largely uncorrelated with income, risk, or credit usage
(Age correlations are near zero across most metrics)	Contrary to expectations, age does not show strong relationships with financial behavior in this dataset. This could be due to a narrow age range in the sample, or simply because other factors (income, occupation, credit access) are more influential.

📌 Including age in modeling may add noise rather than signal.

✅ Actionable: Consider excluding age as a predictor variable unless segmentation is needed by age brackets. It may be more useful in demographic reporting than predictive analytics.
